### **Exploratory Analysis**
"Exploratory Analysis focuses on exploring and understanding the data without preconceived hypotheses. It involves visualizations, summary statistics, and data profiling techniques to uncover patterns, relationships, and interesting features. It helps generate hypotheses for further analysis"

```python
#Exploratory Analysis here

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'])
df['YearMonth'] = df['CRASH DATE'].dt.to_period('M')
crashes_per_month = df.groupby('YearMonth').size()

# Plotting the number of crashes over time (monthly)
plt.figure(figsize=(10, 6))
crashes_per_month.plot(kind='line', marker='o', color='b')

# Adding labels and title
plt.title('Number of Crashes Over Time (Monthly)')
plt.xlabel('Month')
plt.ylabel('Number of Crashes')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Year'] = df['CRASH DATE'].dt.year
crashes_per_year = df.groupby('Year').size()
# Plotting the number of crashes over time (yearly) using a bar plot
plt.figure(figsize=(10, 6))
crashes_per_year.plot(kind='bar', color='g')

# Adding labels and title
plt.title('Number of Crashes Over Time (Yearly)')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')

# Show plot
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Assuming 'CRASH DATE' is a datetime column in the DataFrame
df['Day'] = df['CRASH DATE'].dt.day
```

```python
# Plotting the number of crashes over time (daily) using a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Day'], bins=31, color='orange', edgecolor='black')

# Adding labels and title
plt.title('Number of Crashes Over Time (Daily)')
plt.xlabel('Day of the Month')
plt.ylabel('Frequency')

# Show plot
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Count the number of crashes per borough
borough_counts = df['BOROUGH'].value_counts()

# Plotting the number of crashes per borough
plt.figure(figsize=(10, 6))
borough_counts.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.title('Number of Crashes by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Crashes')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Count the number of crashes per borough
borough_counts = df['BOROUGH'].value_counts()

# Plotting the number of crashes per borough as a pie chart
plt.figure(figsize=(10, 6))
plt.pie(borough_counts, labels=borough_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightyellow'])

# Adding labels and title
plt.title('Distribution of Crashes by Borough')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Count the number of crashes per ZIP code
zip_counts = df['ZIP CODE'].value_counts().head(30)  # Consider top 30 ZIP codes for example

# Plotting the number of crashes per ZIP code
plt.figure(figsize=(10, 6))
zip_counts.plot(kind='bar', color='salmon')

# Adding labels and title
plt.title('Top 10 ZIP Codes with Most Crashes')
plt.xlabel('ZIP Code')
plt.ylabel('Number of Crashes')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Subsetting the DataFrame with the provided columns
selected_columns = [
    'CRASH DATE', 'CRASH TIME', 'BOROUGH', 'ZIP CODE', 'LATITUDE',
    'LONGITUDE', 'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME',
    'OFF STREET NAME', 'NUMBER OF PERSONS INJURED',
    'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
    'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
    'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
    'NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 1',
    'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
    'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
    'COLLISION_ID', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2',
    'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'
]

# Subset the DataFrame with selected columns
df_selected = df[selected_columns]

# Correlation Analysis: Find correlations between numerical columns related to injuries and fatalities
numerical_columns = [
    'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
    'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
    'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
    'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
]

correlation_matrix = df_selected[numerical_columns].corr()

# Frequency Analysis: Count unique values in categorical columns
categorical_columns = [
    'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2',
    'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4',
    'CONTRIBUTING FACTOR VEHICLE 5', 'VEHICLE TYPE CODE 1',
    'VEHICLE TYPE CODE 2', 'VEHICLE TYPE CODE 3',
    'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'
]

frequency_counts = df_selected[categorical_columns].apply(pd.Series.value_counts)

# Missing Values and Data Quality: Analyze missing values and patterns
missing_values = df_selected.isnull().sum()
missing_percentage = (df_selected.isnull().sum() / len(df_selected)) * 100

# Displaying Results
print("Correlation Matrix for Numerical Columns:")
print(correlation_matrix)

print("\nFrequency Counts for Categorical Columns:")
print(frequency_counts)

print("\nMissing Values:")
print(missing_values)

print("\nMissing Values Percentage:")
print(missing_percentage)
```

```python
# Selecting numerical columns related to injuries and fatalities
numerical_columns = [
    'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
    'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
    'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
    'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
]

# Creating a correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'input' is your DataFrame
# Selecting categorical columns for frequency analysis
categorical_columns = [
    'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2',
    'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4',
    'CONTRIBUTING FACTOR VEHICLE 5'
]

# Visualizing frequency counts for categorical columns
plt.figure(figsize=(12, 8))
for col in categorical_columns:
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Frequency Analysis - {col}')
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('')
    plt.ylabel('Count')
    plt.tight_layout()

# Show the plot outside the loop
plt.show()
```

```python
# Visualizing missing values across columns
plt.figure(figsize=(10, 6))
missing_values = df.isnull().sum().sort_values(ascending=False)
sns.barplot(x=missing_values.values, y=missing_values.index, palette='viridis')
plt.title('Missing Values Across Columns')
plt.xlabel('Count of Missing Values')
plt.ylabel('Columns')
plt.tight_layout()
plt.show()
```
